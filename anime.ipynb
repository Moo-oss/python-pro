```python
print(df_cleaned.columns)
df_cleaned.columns = df_cleaned.columns.str.strip().str.lower()
# Ensure 'rating' exists in df_cleaned
if 'rating' in df_cleaned.columns:
    # Sort by 'rating' and select top 10
    top_10_anime = df_cleaned.nlargest(10, 'rating')
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.bar(top_10_anime['title'], top_10_anime['rating'], color='darkblue')
    plt.xlabel('Anime Title')
    plt.ylabel('Rating')
    plt.title('Top 10 Anime by Rating')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("Column 'rating' does not exist in the DataFrame. Please check the dataset.")
```

```python
top_10_anime = df.nlargest(10, 'Rating')

plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_anime, x='Rating', hue='Title', palette='viridis')
plt.title("Top 10 Anime by Rating", fontsize=14)
plt.xlabel("Rating")
plt.ylabel("Title")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], kde=True)  # Use the correct column name: 'Rating'
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['Episodes'], df_cleaned['Rating'], alpha=0.5, color='purple')
plt.title('Rating vs. Number of Episodes')
plt.xlabel('Number of Episodes')
plt.ylabel('Rating')
plt.show()
```

```python
# Line plot: Ratings over time
# Assuming the dataset has a 'date' column in the format 'YYYY-MM-DD'
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'], errors='coerce')
ratings_by_year = df_cleaned.groupby(df_cleaned['date'].dt.year)['rating'].mean().dropna()
plt.figure(figsize=(10, 6))
plt.plot(ratings_by_year.index, ratings_by_year.values, marker='o', linestyle='-', color='c')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Average Anime Rating Over Years')
plt.show()
```

```python
# Bar plot: Number of episodes in top 10 anime
plt.figure(figsize=(10, 6))
plt.bar(top_10_anime['title'], top_10_anime['episodes'], color='brown')
plt.xlabel('Anime Title')
plt.ylabel('Number of Episodes')
plt.title('Number of Episodes in Top 10 Anime')
plt.xticks(rotation=45)
plt.show()
```
